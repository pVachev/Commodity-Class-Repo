```python
# This code is for Session 2 Ex. 1




data_wheat = pd.read_excel("Session2_Exercises.xls", sheet_name="Exercise2", skiprows=18, nrows=5)

data_wheat.drop(data_wheat.index[0], inplace=True)

data_wheat.drop(columns=data_wheat.columns[[0, 5, 6, 7, 8]], inplace=True)

# print(data_wheat.shape)  # Check the number of rows and columns
# print(data_wheat.columns)  # Check the column names

interest_rate_data = pd.read_excel("Session2_Exercises.xls", sheet_name="Exercise2", skiprows=27, usecols="C", nrows=5)

interest_rate_data.columns = ["Interest Rate"]

data_wheat.reset_index(drop=True, inplace=True)
interest_rate_data.reset_index(drop=True, inplace=True)

merged_data = pd.concat([data_wheat, interest_rate_data], axis=1)

def convert_quote(value):
    """Convert a string in the format 506'2 to a float (506.25)"""
    if isinstance(value, str) and "'" in value:
        whole, fraction = value.split("'")
        return float(whole) + float(fraction) / 8  # Convert fraction to decimal
    return float(value)  # If it's already a number, keep it



# Apply conversion
merged_data["Quote"] = merged_data["Quote"].apply(convert_quote)



merged_data["T"] = pd.to_numeric(merged_data["T"], errors='coerce')





merged_data["y"] = np.log((merged_data["Quote"] * np.exp( - merged_data["Interest Rate"] * merged_data["T"])) / 
                          (merged_data["Quote"].shift(1) * np.exp( - merged_data["Interest Rate"].shift(1) * merged_data["T"].shift(1) ))) / (merged_data["T"].shift(1) - merged_data["T"])

merged_data

# merged_data["y"].iloc[0] = 

```
