```python
import pandas as pd 
import numpy as np
```

```python
commodity_swap_data = pd.read_excel("Session3_Exercises.xls", 
                                    nrows=18, index_col=1, skiprows=18)

commodity_swap_data.drop(columns="Unnamed: 0", inplace=True)

commodity_swap_data.drop(index=commodity_swap_data.index[0], inplace=True)

commodity_swap_data["F(0,T)*B(0,T)"] = commodity_swap_data["F(0,T)"] * commodity_swap_data["B(0,T)"]

commodity_swap_data


```

```python
swap_ins_q = pd.DataFrame(np.ones((3,3)), 
                          columns=["Sum of Discount Factors", "Sum of Discounted Future Quotes", "Swap Quote"], 
                          index=["1-Year Monthly Swap", "2-Year Semi-annual Swap","4-Year Annual Swap"])



swap_ins_q
```

```python

swap_ins_q.loc["1-Year Monthly Swap", "Sum of Discount Factors"] = commodity_swap_data["B(0,T)"].head(12).sum()

swap_ins_q.loc["1-Year Monthly Swap","Sum of Discounted Future Quotes"] = commodity_swap_data["F(0,T)*B(0,T)"].head(12).sum()

swap_ins_q["Swap Quote"] = swap_ins_q["Sum of Discounted Future Quotes"] / swap_ins_q["Sum of Discount Factors"]

swap_ins_q
```
